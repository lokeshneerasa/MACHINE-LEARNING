```python
#Ex:12
import numpy as np
from sklearn.model_selection import train_test_split
X=np.arange(20).reshape(2,10).T
y=np.arange(10)
X_train,X_test,y_train,y_test=train_test_split(X,y)
print(X_train)
print()
print(X_test)
print()
print(y_train)
print()
print(y_test)
```

```python
#Ex:13 random_state
import numpy as np
from sklearn.model_selection import train_test_split
X=np.arange(20).reshape(2,10).T
y=np.arange(10)
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=0)
print(X_train)
print()
print(X_test)
print()
print(y_train)
print()
print(y_test)

```
