# **Machine Learning â€“ Pickling and Unpickling**

```python
#Ex:1 using method and class in python
class Employee:
  def __init__(self,eno,ename,esal):
    self.eno=eno
    self.ename=ename
    self.esal=esal

  def display(self):
    print("Number is:",self.eno)
    print("Name is:",self.ename)
    print("salary is:",self.esal)

obj=Employee(533,'Lokesh',54000)
obj.display()
```

```python
#Ex:2 Pickling and unpickling example
import pickle
class Employee:
  def __init__(self,eno,ename,esal):
    self.eno=eno
    self.ename=ename
    self.esal=esal
  def display(self):
    print("Number is:",self.eno)
    print("Name is:",self.ename)
    print("salary is:",self.esal)
with open("emp.data","wb") as f:
  e=Employee(533,'Lokesh',55000)
  pickle.dump(e,f)
  print("pickling of employee object completed\n")
with open("emp.data","rb") as f:
  obj=pickle.load(f)
  print("unpickling of employee information completed\n")
  obj.display()
```

```python
#Ex:2 Save the  model into pickle file
import pandas as pd
from sklearn.linear_model import LinearRegression
import pickle

df=pd.read_csv("homeprices.csv")

new_df=df.drop('price',axis='columns')

model=LinearRegression()
model.fit(new_df.values,df.price.values)

with open('model_pickle','wb') as file:
  pickle.dump(model,file)

with open('model_pickle','rb') as file:
  model1=pickle.load(file)
  print(model1.predict([[5000]]))
```
